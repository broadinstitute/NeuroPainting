```python
# Batch 15 dataset
df = pd.read_csv(
    'C:/Users/Ruifan/Downloads/high_content_imaging_cellprofiler_Results_Batch15_Lvesi45132021_BackendFiles_210512_151243-V_210512_151243-V_normalized.csv',
    index_col=0)
cols = df.columns.tolist()
cols = [i for i in cols if i.find('Costes')<0 if i.find('Manders')<0 if i.find('RWC')<0]
df = df[cols]
df = df.dropna(axis=1, how='any')
groups = df['Metadata_Peturb_gene'].tolist()
```

```python
%%time
df_noise_removed = feature_select(df, operation='noise_removal', perturb_list=groups, stdev_cutoff=0.8)
```

```python
df = df[[i for i in df.columns if 'Metadata' not in i]]
df['group'] = groups

perturb_list = list(set(df['group'].tolist()))
feature_list = df.drop('group', axis=1).columns.tolist()
features_std_to_keep = []

keep_threshold = 0.8 
mean_stdevs = []

for feature in feature_list:
    feature_stdevs = [] # list of stdevs per feature, should equal number of perturbations
    for perturb in perturb_list:
        temp = df[df['group'] == perturb][feature]
        temp_values = list(temp.values)
        feature_stdevs.append(np.std(temp_values))
    feature_mean_stdev = np.mean(feature_stdevs)
    mean_stdevs.append(feature_mean_stdev)
    if feature_mean_stdev < keep_threshold:
        features_std_to_keep.append(feature)
    # Reset feature stdevs for the next feature
#     feature_stdevs = []
    
print(len(features_std_to_keep))
```

```python
df_noise_removed = df_noise_removed[[i for i in df_noise_removed.columns if 'Metadata' not in i]]
df_noise_removed['group'] = groups

perturb_list = list(set(df_noise_removed['group'].tolist()))
feature_list = df_noise_removed.drop('group', axis=1).columns.tolist()
features_std_to_keep = []

keep_threshold = 0.8 
mean_stdevs2 = []

for feature in feature_list:
    feature_stdevs = [] # list of stdevs per feature, should equal number of perturbations
    for perturb in perturb_list:
        temp = df_noise_removed[df_noise_removed['group'] == perturb][feature]
        temp_values = list(temp.values)
        feature_stdevs.append(np.std(temp_values))
    feature_mean_stdev = np.mean(feature_stdevs)
    mean_stdevs2.append(feature_mean_stdev)
    if feature_mean_stdev < keep_threshold:
        features_std_to_keep.append(feature)
    # Reset feature stdevs for the next feature
#     feature_stdevs = []
    
print(len(features_std_to_keep))
```

```python
fig, ax = plt.subplots(figsize=(12,5))
sns.distplot(mean_stdevs, ax=ax)
sns.distplot(mean_stdevs2, ax=ax)
plt.xlabel('StDev')
# plt.xlim(0,4)
```
